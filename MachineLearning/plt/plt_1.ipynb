```python
import matplotlib.pyplot as plt
import numpy as np
```

# plot

```python
x=np.linspace(-1,1,50)
y=2*x+1
plt.plot(x,y)
plt.show()
```

# figure

```python
x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2


plt.figure(num=1)# 从这里开始是第一个figure窗口里的内容
plt.plot(x,y1)

plt.figure(num=2,figsize=(8,5))#figsize可以设置长宽
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')# 颜色，宽度，线条格式
```

# axis

```python
plt.figure(num=3,figsize=(8,5))#figsize可以设置长宽
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')# 颜色，宽度，线条格式

plt.xlim((-1,2))
plt.ylim((-2,3))# x，y轴范围
plt.xlabel('x-label')
plt.ylabel('y-label')
```

```python
# 更换间距tick
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')

plt.xlim((-1,2))
plt.ylim((-2,3))# x，y轴范围
plt.xlabel('x-label')
plt.ylabel('y-label')

new_ticks=np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)
```

# legend图例

```python
plt.figure(figsize=(10,6))
plt.plot(x,y2,label='up')
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--',label='down')# 颜色，宽度，线条格式
plt.legend(loc='lower right')

'''
best
upper right
upper left
lower left
lower right
right
center left
center right
lower center
upper center
center
'''
```

# Annotation标注

```python
plt.figure(figsize=(10,6))
plt.plot(x,y2,label='up')
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--',label='down')# 颜色，宽度，线条格式
plt.legend(loc='lower right')

x0=1
y0=2*x0+1
plt.scatter(x0,y0,s=50,color='b')#blue

plt.plot([x0,x0],[y0,-5],'k--',lw=2.5)# 画一条虚线
```

# scatter 散点图

```python
n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)# normal 正态分布
T=np.arctan2(Y,X)

plt.scatter(X,Y,s=75,c=T,alpha=0.5)# alpha透明度50%
plt.xlim((-1.5,1.5))#注意双括号
plt.ylim((-1.5,1.5))
```
