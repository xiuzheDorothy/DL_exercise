```python
import matplotlib.pyplot as plt
import numpy as np
```

# plot

```python
x=np.linspace(-1,1,50)
y=2*x+1
plt.plot(x,y)
plt.show()
```

# figure

```python
x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2


plt.figure(num=1)# 从这里开始是第一个figure窗口里的内容
plt.plot(x,y1)

plt.figure(num=2,figsize=(8,5))#figsize可以设置长宽
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')# 颜色，宽度，线条格式
```

# axis

```python
plt.figure(num=3,figsize=(8,5))#figsize可以设置长宽
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')# 颜色，宽度，线条格式

plt.xlim((-1,2))
plt.ylim((-2,3))# x，y轴范围
plt.xlabel('x-label')
plt.ylabel('y-label')
```

```python
# 更换间距tick
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')

plt.xlim((-1,2))
plt.ylim((-2,3))# x，y轴范围
plt.xlabel('x-label')
plt.ylabel('y-label')

new_ticks=np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)
```

# legend图例

```python
plt.figure(figsize=(10,6))
plt.plot(x,y2,label='up')
plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--',label='down')# 颜色，宽度，线条格式
plt.legend(loc='lower right')

'''
best
upper right
upper left
lower left
lower right
right
center left
center right
lower center
upper center
center
'''
```
