## Tick Label Formats 轴标签的格式

x,y轴会有各种各样的标注形式，比如以log为单位，以e为单位，以百分比为但单位。

bokeh有一种万物皆为对象的思想，因而以上也都是用对象实现的

- BasicTickFormatter — Default formatter for linear axes.
- CategoricalTickFormatter — Default formatter for categorical axes.
- DatetimeTickFormatter — Default formatter for datetime axes.
- LogTickFormatter — Default formatter for log axes.

源码点这里[here](https://docs.bokeh.org/en/latest/_modules/bokeh/models/formatters.html#BasicTickFormatter)

一个简单的例子：bokeh中的数据如果大了就会用科学计数法表示，我们将其变化为普通的表示法。

> 另外：pandas不用科学计数的表示可以直接设置：
```python
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

````python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook,show
from bokeh.plotting import figure
output_notebook() #导入这个命令之后bokeh的绘画内容都在notebook上显示

pd.set_option('display.float_format',lambda x : '%.2f' % x)
````

````python
excel=pd.ExcelFile('data_AZ_TETCB.xlsx')
data=pd.read_excel(excel,'Sheet1')
p=figure(plot_width=400, plot_height=400)
p.line(data['Year'],data['Data'])
show(p)
````

原图可见这是科学计数法

````python
from bokeh.models import NumeralTickFormatter
p2=figure(plot_width=400, plot_height=400)
p2.line(data['Year'],data['Data'])

p2.yaxis[0].formatter=NumeralTickFormatter()# 作为实例化对象引入，最好新建一个对象
#p2.yaxis[0].formatter.use_scientific==False

#p1.line(data['Year'],data['Data'])
show(p2)
````

很奇怪的一点，按理说拒绝bokeh列显示为科学计数法是要调用`BasicTickFormatter`类的，因为在它的源码里面写明了
```python
 use_scientific = Bool(True, help="""
    Whether to ever display scientific notation. If ``True``, then
    when to use scientific notation is controlled by ``power_limit_low``
    and ``power_limit_high``.
    """)
```
意思只要设置`Flase`就好了啊，但是亲测不行。所以改为引入`NumeralTickFormatter`并且不附带参数，因为它是基于字符串的：Tick formatter based on a human-readable format string，不加参数直接实例化的结果就是上述的普通的计数法
