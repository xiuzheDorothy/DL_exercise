## 手工制作数据透视表
先用groupby计算一下分性别的生还率

```python
titanic.groupby('sex')[['survived']].mean()
```

但这还不够，我们想要一个二维的可自由移动的图表，`groupby`只有一维的，或许可以这样：

```python
 titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

但这样非常麻烦。而用数据透视表`pivot_table`可以轻松解决这个问题

## 数据透视表语法

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

### 多维数据透视表
这里尝试把age加进去作为第三个维度

```python
age=pd.cut(titanic['age'],[0,18,80])# 分成两段
titanic.pivot_table('survived',['sex',age],'class')
```
